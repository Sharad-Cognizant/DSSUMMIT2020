```python
import numpy as np
import random
```

```python
class EightQueenSolution():
    
    def __init__(self,populationLimit):
        self.populationLimit = populationLimit
        self.populationList = []
        self.populationDict = {}
        self.fitnessScoreDict = {}
        self.solution = []
        
    def initiateBoard(self):
        self.board = np.zeros(shape=(8,8),dtype = int)  
        return self.board
        
    def genPopulation(self):
        
        for i in range(self.populationLimit):
            self.len = 0
            self.chromosome = []
            while self.len<8:
                j = random.randint(0,7)
                if j not in self.chromosome:
                    self.chromosome.append(j)
                    self.len = self.len+1

            self.populationList.append(self.chromosome)
            self.populationDict[i] = self.chromosome
        
        return self.populationList
    
    
    def populateBoardandFitnessCalculation(self):
        
        self.populationElement = 0
        for self.chromosome in self.populationList:
            self.generatedBoard = self.initiateBoard()
            self.col = 0
            for self.DNA in self.chromosome:
                self.generatedBoard[self.DNA][self.col] = 1
                self.col = self.col+1
            self.fitnessScore = self.calculateFitness(self.generatedBoard)
            self.fitnessScoreDict[self.populationElement] = self.fitnessScore
            self.populationElement = self.populationElement + 1

        return self.fitnessScoreDict
    
    
    def calculateFitness(self,board):
        #Diagonal1
        #Lower Diagonal 1
        self.board = board

        self.counter = 0
        self.hits = 0
        self.row=0
        self.col = 0

        while self.counter<=8:
            self.flag = 0
            self.row = self.counter
            for self.col in range(0,8-self.counter):
                if self.board[self.row][self.col]== 1:
                    self.flag = self.flag + 1
                    if self.flag > 1:
                        self.hits = self.hits + 1
                #print(row,col)
                self.row = self.row + 1
            self.counter = self.counter + 1

        #Upper Diagonal 1

        self.counter1 = 1

        while (self.counter1<=8):
            self.flag = 0
            self.col1 = self.counter1
            for self.row in range(0,8-self.counter1):
                if self.board[self.row][self.col1]== 1:
                    self.flag = self.flag + 1
                    if self.flag > 1:
                        self.hits = self.hits + 1
                #print(row,col1)
                self.col1 = self.col1 + 1
            self.counter1 = self.counter1 + 1

        #Diagonal 2
        #Lower Diagonal 2

        self.counter = 0
        while self.counter<=8:
            self.flag = 0
            self.col = self.counter
            for self.row in range(7,self.counter-1,-1):
                if self.board[self.row][self.col]== 1:
                    self.flag = self.flag + 1
                    if self.flag >1:
                        self.hits = self.hits + 1
                #print(row,col)
                self.col = self.col + 1
            self.counter = self.counter + 1

        #Upper Diagonal 2

        self.counter = 6
        while self.counter>=0:
            self.flag = 0
            self.row = self.counter
            for self.col in range(0,self.counter+1):
                if self.board[self.row][self.col]== 1:
                    self.flag = self.flag + 1
                    if self.flag > 1:
                        self.hits = self.hits + 1
                #print(row,col)
                self.row = self.row - 1
            self.counter = self.counter - 1

        #checking row wise    
        for self.row in range(0,8):
            self.flag = 0
            for self.col in range(0,8):
                if self.board[self.row][self.col] == 1:
                    self.flag = self.flag + 1
                    if self.flag > 1:
                        self.hits = self.hits + 1
                #print(row,col)

        for self.col in range(0,8):
            self.flag = 0
            for self.row in range(0,8):
                if self.board[self.row][self.col] == 1:
                    self.flag = self.flag + 1
                    if self.flag > 1:
                        self.hits = self.hits + 1
                #print(row,col)

        return self.hits
    
    
    
    
    def findFitParents(self, fitnessScoreDicSortedByValue):
        self.fittestParents = []
        self.fitnessScoreDicSortedByValue = fitnessScoreDicSortedByValue
        self.count = 0
        for self.i,self.j in self.fitnessScoreDicSortedByValue:
            if self.count<=1:
                #print(i)
                #print(populationList[i])
                self.fittestParents.append(self.populationList[self.i])
                self.count = self.count + 1
            else:
                break

        return self.fittestParents
    
    
    def crossOver(self,fittestParents):
        
        self.fittestParents = fittestParents
        
        self.parent1 = self.fittestParents[0]
        self.parent2 = self.fittestParents[1]
        self.child = [0,0,0,0,0,0,0,0]
        self.count = 0
        for i in range(7):
            if self.count <=3:
                self.child[i] = self.parent1[i]
                self.child[7-i] = self.parent2[7-i]
                self.count = self.count + 1
            else:
                break

        return self.child 
    
    
    def mutation(self, child):
        self.child = child
        self.index = random.randint(0,7)
        while True:
            self.value = random.randint(0,7)
            if self.value not in self.child:
                self.child[self.index] = self.value
                break 
        #print('Mutated Child')
        #print(child)

        #populationList.append(child)

        return self.child
    
    def appendMutatedChild(self,mutatedChild):
        self.populationList.append(mutatedChild)
        
        return self.populationList
    
    def GenerateSolution(self,fitnessScoreDicSortedByValue):
        
        self.fitnessScoreDicSortedByValue = fitnessScoreDicSortedByValue
        
        for self.index, self.score in self.fitnessScoreDicSortedByValue:
            #print(index,score)
            if self.score == 0:
                self.solution.append(self.populationList[self.index])
                
        print(self.solution[0])
    
   # def populateBoard(self):
        
        #print ("Solutions in list are")
        #print(self.solution)
        #print('Solutions Boards are below')
        #populationElement = 0
        #self.i = 1
#         for self.chromosome in self.solution:
#             print(f"Solution {self.i}")
#             print(self.chromosome)
#             self.generatedBoard = self.initiateBoard()
#             self.col = 0
#             for self.DNA in self.chromosome:
#                 self.generatedBoard[self.DNA][self.col] = 1
#                 self.col = self.col+1
#             print("Queens Positions generated on Board for above Solution:-")
#             print(self.generatedBoard)
#             print('**'*20)
#             self.i = self.i + 1
            
        
        
       
        
    

```
